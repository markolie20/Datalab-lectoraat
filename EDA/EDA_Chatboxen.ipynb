```python
def plot_group_distribution_percent(df, column, top_n=100, sort_by_value=None):
    # Bereken % distributie per groep
    df_counts = (
        df.groupby(['group', column])
        .size()
        .groupby(level=0)
        .apply(lambda x: 100 * x / float(x.sum()))
        .unstack()
        .fillna(0)
    )

    # Alleen top N groepen op basis van groepsgrootte
    top_groups = df['group'].value_counts().head(top_n).index
    df_counts = df_counts.loc[top_groups]

    # Sorteer op specifieke waarde (bijv. 'male')
    if sort_by_value and sort_by_value in df_counts.columns:
        df_counts = df_counts.sort_values(by=sort_by_value, ascending=False)

    # Plot zonder randen tussen de balken
    df_counts = df_counts.sort_values('M', ascending=False)
    ax = df_counts.plot(kind='bar', stacked=True, colormap='tab10', edgecolor='none', width=1)

    ax.set_title(f"Gestapelde procentuele verdeling van '{column}' (top {top_n} groepen, gesorteerd op '{sort_by_value}')")
    ax.set_ylabel("Percentage")
    ax.set_xticks([])  # Verberg x-ticks
    ax.legend(title=column, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
    
    return df_counts


tst = plot_group_distribution_percent(df, 'sex', top_n=1000, sort_by_value='male')
```

```python
import matplotlib.pyplot as plt

def plot_stack_area_chart(df, column, top_n=100):
    # Zorg ervoor dat we werken met percentage distributie per groep
    df_counts = (
        df.groupby(['group', column])
        .size()
        .groupby(level=0)
        .apply(lambda x: 100 * x / float(x.sum()))
        .unstack()
        .fillna(0)
    )

    # Alleen top N groepen op basis van groepsgrootte
    top_groups = df['group'].value_counts().head(top_n).index
    df_counts = df_counts.loc[top_groups]

    # Plotten als gestapelde gebiedsdiagram
    df_counts = df_counts.sort_values(df_counts.columns[0] ,ascending = False)
    ax = df_counts.plot(kind='area', stacked=True, figsize=(12, 6), colormap='tab10')

    ax.set_title(f"Gestapelde gebiedsdiagram van '{column}' (top {top_n} groepen)")
    ax.set_ylabel("Percentage")
    ax.set_xlabel("Groep")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return df_counts
tst = plot_stack_area_chart(df, 'household_composition', 1000)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

def plot_aggregated_stack_area(
    df: pd.DataFrame,
    group_col: str,          # Column name for groups (e.g., 'group')
    category_col: str,       # Column name for categories (e.g., 'household_composition')
    top_n: int = 1000,       # Number of top groups (by count) to consider
    num_bins: int = 20,      # How many bins to aggregate into
    sort_by_category: str = None, # Which category percentage to sort by before binning
    sort_ascending: bool = False, # Sort order (False = descending)
    figsize: tuple = (12, 6),
    colormap: str = 'tab10'
):
    """
    Generates and displays an aggregated stacked area chart.

    Calculates category percentages within groups, filters for top N groups,
    sorts them by a specified category's percentage, aggregates into bins,
    and plots the average composition per bin.

    Args:
        df (pd.DataFrame): Input DataFrame with raw data.
        group_col (str): Name of the column containing group identifiers.
        category_col (str): Name of the column containing the categories to stack.
        top_n (int, optional): Consider only the top N groups based on their
                               total count in `df`. Defaults to 1000.
        num_bins (int, optional): Number of bins to aggregate the sorted groups
                                  into for plotting. Defaults to 20.
        sort_by_category (str, optional): The category name (value within
                                          `category_col`) whose percentage
                                          should be used for sorting groups
                                          before binning. If None, defaults to
                                          the first category found. Defaults to None.
        sort_ascending (bool, optional): Sort order for `sort_by_category`.
                                         Defaults to False (descending).
        figsize (tuple, optional): Figure size. Defaults to (12, 6).
        colormap (str, optional): Colormap for the plot. Defaults to 'tab10'.

    Returns:
        pd.DataFrame: The aggregated DataFrame used for plotting (average
                      percentages per bin). Returns None if plotting fails.

    Raises:
        ValueError: If specified columns or sort category are not found, or
                    if num_bins is invalid.
    """
    # --- Input Validation ---
    if group_col not in df.columns:
        raise ValueError(f"Group column '{group_col}' not found in DataFrame.")
    if category_col not in df.columns:
        raise ValueError(f"Category column '{category_col}' not found in DataFrame.")

    # --- 1. Calculate Percentage Distribution per Group ---
    try:
        df_percentages = (
            df.groupby([group_col, category_col])
            .size()
            .groupby(level=0) # Group by group_col index
            .apply(lambda x: 100 * x / float(x.sum()))
            .unstack(fill_value=0) # Pivot categories to columns, fill missing with 0
        )
    except ZeroDivisionError:
        print("Warning: Some groups might have zero total counts, leading to division by zero.")
        # Handle potentially empty groups if necessary, e.g., by filtering them out first.
        # For now, we proceed, unstack(fill_value=0) handles groups with some categories missing.
        # Re-calculate potentially filtering groups with zero total counts if it causes issues.
        group_counts = df.groupby(group_col).size()
        valid_groups = group_counts[group_counts > 0].index
        df_filtered = df[df[group_col].isin(valid_groups)]
        if df_filtered.empty:
             print("Error: No valid groups with counts > 0 found.")
             return None
        df_percentages = (
            df_filtered.groupby([group_col, category_col])
            .size()
            .groupby(level=0)
            .apply(lambda x: 100 * x / float(x.sum()))
            .unstack(fill_value=0)
        )


    if df_percentages.empty:
        print("Warning: Percentage DataFrame is empty after calculations.")
        return None

    # --- 2. Filter for Top N Groups (Optional) ---
    if top_n is not None and top_n < len(df_percentages):
        # Get top N groups based on total count in original df
        top_groups_by_count = df[group_col].value_counts().head(top_n).index
        # Filter the percentage dataframe - use intersection for safety
        valid_top_groups = df_percentages.index.intersection(top_groups_by_count)
        df_percentages = df_percentages.loc[valid_top_groups]
        if df_percentages.empty:
             print(f"Warning: No groups remaining after filtering for top {top_n}.")
             return None


    actual_num_groups = len(df_percentages)
    if actual_num_groups == 0:
        print("Warning: DataFrame is empty after filtering. Cannot generate plot.")
        return None

    # --- 3. Sort Groups ---
    actual_categories = df_percentages.columns.tolist()
    if sort_by_category is None:
        sort_by_category = actual_categories[0] # Default to first column
        print(f"Warning: No 'sort_by_category' specified. Sorting by first category found: '{sort_by_category}'")
    elif sort_by_category not in actual_categories:
        raise ValueError(f"Category '{sort_by_category}' specified for sorting not found "
                         f"in calculated percentage columns: {actual_categories}")

    df_sorted = df_percentages.sort_values(by=sort_by_category, ascending=sort_ascending)
    # Reset index here so we can assign bins based on row number easily
    df_sorted = df_sorted.reset_index(drop=True)


    # --- 4. Create Bins and Aggregate ---
    num_bins = min(num_bins, actual_num_groups) # Cannot have more bins than groups
    if num_bins <= 0:
        raise ValueError("Number of bins must be positive.")

    groups_per_bin = actual_num_groups // num_bins

    # Create an array indicating which bin each row (sorted group) belongs to
    bin_assignments = np.repeat(np.arange(num_bins), groups_per_bin)

    # Distribute remainder groups more evenly across the last bins
    remainder = actual_num_groups % num_bins
    if remainder > 0:
        extras = np.repeat(np.arange(num_bins - remainder, num_bins), 1)
        bin_assignments = np.concatenate([bin_assignments, extras])

    # Ensure bin_assignments has the correct length (important if remainder was 0)
    bin_assignments = bin_assignments[:actual_num_groups]

    df_sorted['bin'] = bin_assignments

    # Calculate the average percentage for each original category within each bin
    df_agg = df_sorted.groupby('bin')[actual_categories].mean()

    # --- 5. Create Bin Labels ---
    bin_labels = []
    start_idx = 0
    for i in range(num_bins):
        # Find the last index (in the sorted df) belonging to the current bin
        end_idx = df_sorted[df_sorted['bin'] == i].index.max()
        if pd.isna(end_idx): continue # Skip if a bin ended up empty

        # Label includes bin number and range of sorted indices it covers
        bin_labels.append(f"Bin {i+1}\n(Sorted Idx {start_idx}-{int(end_idx)})")
        start_idx = int(end_idx) + 1

    # --- 6. Plot the Aggregated Data ---
    fig, ax = plt.subplots(figsize=figsize)
    df_agg.plot(kind='area', stacked=True, ax=ax, colormap=colormap)

    # --- 7. Customize Plot ---
    sort_order_desc = "Descending" if not sort_ascending else "Ascending"
    title = (f"Aggregated Stacked Area Chart of '{category_col}' ({num_bins} Bins)\n"
             f"({actual_num_groups} Groups from Top {top_n if top_n else 'All'}, "
             f"Sorted by % {sort_by_category} {sort_order_desc})")
    ax.set_title(title)
    ax.set_ylabel("Average Percentage (%)")
    ax.set_xlabel("Group Bins (based on sorted order)")

    # Set x-axis ticks and labels
    #ax.set_xticks(np.arange(len(bin_labels))) # Use number of labels generated
    ax.set_xticklabels(bin_labels, rotation=45, ha='right') # Apply bin labels

    # Set y-axis limits and format
    ax.set_ylim(0, 100)
    ax.yaxis.set_major_formatter(mticker.PercentFormatter(xmax=100.0))

    # Add Legend
    ax.legend(title=category_col, loc='center left', bbox_to_anchor=(1, 0.5))

    # Adjust layout
    plt.tight_layout(rect=[0, 0, 0.85, 1]) # Make space for legend
    plt.show()

    return df_agg

```

```python
col_list = ['country_of_origin', 'socio_economic_status', 'household_composition',
            'political_orientation', 'education_level', 'religion', 'marital_status', 'employment_status',
            'housing_type', 'technology_proficiency']
for col_name in col_list:
    plot_aggregated_stack_area(
    df=df,
    group_col='group',                 # Your column with group names
    category_col=col_name, # Your column with categories
    top_n=1000,                        # How many top groups to analyze
    num_bins=25,                       # How many bins for the final plot x-axis
    sort_ascending=False               # Show highest % of 'Alleenstaand' on the left
)
```

```python
import os
import json
import matplotlib.pyplot as plt
import numpy as np

# Pad naar je JSON-map
data_dir = r'data\PD Backend en AI-applicatie – Documentatie en Testdata\bios'  # <-- vervang dit

bios_per_file = {}

# Data verzamelen
for filename in sorted(os.listdir(data_dir)):
    if filename.endswith('.json'):
        filepath = os.path.join(data_dir, filename)
        with open(filepath, 'r', encoding='utf-8') as f:
            data = json.load(f)
            bios_per_file[filename] = len(data)

# Voor de plot
file_names = list(bios_per_file.keys())
counts = list(bios_per_file.values())
x = np.arange(len(file_names))

# Selecteer indices voor ticks (0%, 25%, 50%, 75%, 100%)
tick_indices = [0, len(x) // 4, len(x) // 2, 3 * len(x) // 4, len(x) - 1]
tick_labels = [file_names[i] for i in tick_indices]

# Plotten
plt.figure(figsize=(12, 6))
plt.bar(x, counts)
plt.xticks(tick_indices, tick_labels, rotation=45)
plt.xlabel("JSON-bestanden")
plt.ylabel("Aantal bios")
plt.title("Aantal bios per JSON-bestand")
plt.tight_layout()
plt.show()
```

```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from nltk.corpus import stopwords

# ---- 1. Load alle bio's uit JSON-bestanden ----
all_bios = []
names = []

for filename in os.listdir(data_dir):
    if filename.endswith('.json'):
        with open(os.path.join(data_dir, filename), 'r', encoding='utf-8') as f:
            data = json.load(f)
            for person in data:
                names.append(person.get("Name", "Onbekend"))
                all_bios.append(person.get("bio", ""))

# ---- 2. TF-IDF vectorisatie ----
nltk.download('stopwords')

# Laad Nederlandse stopwoorden
dutch_stops = list(stopwords.words('dutch'))  # ← fix: maak er een lijst van
vectorizer = TfidfVectorizer(max_features=500, stop_words=dutch_stops)

X = vectorizer.fit_transform(all_bios)
feature_names = np.array(vectorizer.get_feature_names_out())

# ---- 3. PCA (2D reductie) ----
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())

# ---- 4. KMeans clustering ----
n_clusters = 4  # <- pas dit aan als je wilt
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(X_pca)

# ---- 5. Visualisatie ----
plt.figure(figsize=(10, 8))
for cluster_id in range(n_clusters):
    plt.scatter(
        X_pca[labels == cluster_id, 0],
        X_pca[labels == cluster_id, 1],
        label=f"Cluster {cluster_id + 1}",
        alpha=0.6
    )
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("PCA + KMeans Clustering op Bio's")
plt.legend()
plt.grid(True)
plt.show()

# ---- 6. Topwoorden per PCA-component ----
top_n = 10  # hoeveel woorden per component
print("\nTopwoorden per PCA-component:")

for i, component in enumerate(pca.components_):
    top_indices = np.argsort(component)[-top_n:]
    top_words = feature_names[top_indices]
    print(f"\n🔹 Component {i+1}: {', '.join(top_words)}")
```
